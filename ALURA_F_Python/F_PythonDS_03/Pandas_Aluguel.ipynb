# Relatório de Análise IV

## Seleções e Frequências

```python
dados = pd.read_csv('aluguel_residencial_semindice.csv', sep = ';')
```

```python
dados.head()
```

### Seleção de Imóveis classificados com Apartamento

```python
# Fazendo a seleção dos dados do tipo apartamento
selecao1 = dados['Tipo'] == 'Apartamento'

# Filtrando no dataset os dados selecionados
n1 = dados[selecao1].shape[0]

# Visualizando a frequência 
n1
```

### Seleção de imóveis classificados com tipo Casa , Casa Condomínio e Casa Vila

```python
# Fazendo a seleção dos tipos desejados, como é mais de um, usamos a barra vertical '|' que representa o operador lógico OR
selecao2 = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa Vila')

# Filtrando no dataset os dados selecionados
n2 = dados[selecao2].shape[0]

# Visualizando a frequência 
n2
```

### Seleção de imóveis com área entre 60 a 100 metros quadrados, incluindo os limite

```python
# notação matemática seria 60 <= Área <= 100
# Fazendo a seleção dos dados pela área desejada
selecao3 = (dados['Area'] >= 60) & (dados['Area'] <= 100)

# Filtrando no dataset os dados selecionados
n3 = dados[selecao3].shape[0]

# Visualizando a frequência 
n3
```

### Seleção de imóveis com 4 quartos ou mais e aluguel menor que R$ 2.000,00

```python
# Fazendo a seleção dos dados filtrando pela quantidade de Quartos e pelo Valor
selecao4 = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)

# Filtrando no dataset os dados selecionados
n4 = dados[selecao4].shape[0]

# Visualizando a frequência 
n4
```

```python
print("Nº de Imóveis classificados com Apartamento: {}".format(n1))
print("Nº de imóveis classificados com tipo Casa , Casa Condomínio e Casa Vila: {}".format(n2))
print("Nº de imóveis com área entre 60 a 100 metros quadrados, incluindo os limite: {}".format(n3))
print("Nº de imóveis com 4 quartos ou mais e aluguel menor que R$ 2.000,00: {}".format(n4))
```

# Relatório de Análise V

## Tratamento de Dados Faltantes (missing values)

```python
dados = pd.read_csv('aluguel_residencial_semindice.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# métodos que nos ajudam a visualizar os valores ausentes

dados.isnull()
```

```python
dados.notnull()
```

```python
# mas ainda assim esses métodos não nos ajudam a visualizar de forma efetiva
# usando o metodo info() para visualizar os dados de um df

dados.info()

# com ela já conseguimos identificar que a variável Valor, Condominio e IPTU tem valore ausentes
```

```python
# visualizando os valores nulos

dados[dados['Valor'].isnull()]
```

```python
# para essa variável optaremos por eliminar esses registros com o método dropna

A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
dados[dados['Valor'].isnull()]
```

```python
# Vamos assumir que todo apartamento tem que ter um condominio, com isso estabelecido, \
# vamos eliminar todos os registros nulos que não atenderem essa premissa

dados[dados['Condominio'].isnull()].shape[0]
```

```python
# selecionando os registros dos apartamentos sem valor de condominio
selecao5 = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# fazendo remoção
A = dados.shape[0]
dados = dados[~selecao5] # colocamos o '~' na frente da seleção para que ele inverta a seleção 
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# nesse caso vamos manter esses registros no dataset, mas vamos atribuir o valor 0 a eles

# dados.fillna(0, inplace = True) # primeiro possibilidade

dados = dados.fillna({'Condominio': 0, 'IPTU': 0}) # segunda possibilidade (ela permite passar valores diferentes para cada variável)
dados
```

```python
dados.info()
```

```python
# Salvando dataset com alterações

dados.to_csv('aluguel_limpo.csv', sep = ';', index = False)
```

# Relatório de Análise VI

## Criando Novas Variáveis

```python
dados = pd.read_csv('aluguel_limpo.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# Criando a primeira variável: Valor Bruno

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
# Criando a segunda variável: Valor m²

dados['Valor M²'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
# formatando as casas decimais
dados['Valor M²'] = dados['Valor M²'].round(2)
dados.head(10)
```

```python
# Criando a terceira variável: Valor Bruto M² e ja formalizando

dados['Valor Bruto M²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
# Criando uma nova variável para identificar os imóveis como CASA ou Apartamento
# função apply permite que a gente aplique uma função em casa registro
# no caso usaremos uma função lambda com uma função condicional (if)

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(30)
```

### Excluindo variáveis

```python
# Devido aos valores de IPTU não serem muito confiáveis e acabar comprometendo o valor bruto, vamos excluir 
# as variáveis Valor bruto e valor bruto m²

dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor Bruto', 'Valor Bruto M²', 'Valor M²']])
```

```python
dados_aux
```

```python
# primeira forma de excluir uma variável de um dataframe

del dados_aux['Valor Bruto']
dados_aux
```

```python
# segunda forma de excluir uma variavel de um dataframe

dados_aux.pop('Valor Bruto M²')
dados_aux
```

```python
# terceira forma de excluir uma variável de um dataframe
# a função drop funciona tanto para linha quanto coluna, no caso como queremos excluir a coluna, vamos passar 
# o eixo (axis = 1 quer dizer coluna)

dados.drop(['Valor Bruto', 'Valor Bruto M²'], axis=1, inplace=True)
dados.head(15)
```

```python
dados.to_csv('aluguel_v5.csv', sep = ';', index= False)
```

```python
dados.to_csv('extras/aluguel_v5.csv', sep = ';', index= False)
```

# Relatório de Análise VIII

## Identificando e Removendo Outliers

```python
# usando matplotlib para visualizações gráficas
# config para mostrar as visualizações no jupyter 

# inline mostrara um gráfico estática na tela
%matplotlib inline

import matplotlib.pyplot as plt

# configura o tamanho que queremos que essa figura fique na tela
plt.rc('figure', figsize = (14,6))
```

```python
dados = pd.read_csv('aluguel_v5.csv', sep = ';')
```

```python
# Para remoção dos outliers, vamos trabalhar com o boxplot

dados.boxplot(['Valor'])

# É possível ver que temos dados com muitos problemas, por isso não é possível visualizar os dados que estão entre o Q1, Q3 e os limites
```

```python
# Conforme é percebido no gráfico, esses outliers estão acima de 500000, então vamos visualizar quem são

dados[dados['Valor'] >= 500000]
```

```python
# para facilitar a digitação, vamos criar uma series

valor = dados['Valor']
valor
```

<img src="box-plot.png" width=70%>

### Calculando as estatisticas

```python
# quartis,  intervalo interquartil. limite inferior e limite superior
# quantile() função para calcular os quartils (0.25, 0.50, 0.75)

Q1 = valor.quantile(0.25) 
Q3 = valor.quantile(0.75)
IIQ = Q3 - Q1
LIMINF = Q1 - 1.5 * IIQ
LIMSUP = Q3 + 1.5 * IIQ


print(Q1,Q3, IIQ, LIMINF, LIMSUP)
```

```python
# faremos uma seleção dos dados que estão dentro do limite e o que estiver fora, os outliers, seram descartados

selecao7 = (valor >= LIMINF) & (valor <= LIMSUP)

# vamos criar uma nova variavel para não danificar nosso DF
dados_semoutliers = dados[selecao7]
```

```python
# criando um novo boxplot para visualizar se os dados continuam achatados

dados_semoutliers.boxplot(['Valor'])
```

```python
dados_semoutliers
```

```python
# vamos visualizar a distribuição de frequencia com um histograma

dados_semoutliers.hist(['Valor'])
```
