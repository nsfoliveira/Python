# Concatenando dataframes

```python
listas = [[1,2,3,], 
        [4,5,6,], 
        [7,8,9]]
```

```python
listas
```

```python
df1 = pd.DataFrame(data = listas)
```

```python
df1
```

```python
rot_linhas = ['Linha ' + str(i) for i in range(3)]
```

```python
rot_linhas
```

```python
df1 = pd.DataFrame(data = listas, index = rot_linhas)
```

```python
df1
```

```python
    rot_col = ['Coluna ' + str(i) for i in range(3)]
```

```python
df1 = pd.DataFrame(data = listas, index = rot_linhas, columns = rot_col)
```

```python
df1
```

```python
listas2 = [(1,2,3),
          (4,5,6,),
          (7,8,9)]
```

```python
df2 = pd.DataFrame(data = listas2, index = rot_linhas, columns = rot_col)
```

```python
df1[df1 > 0] = 'A'
df2[df2 > 0] = 'B'
```

```python
df1
```

```python
df2
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
df3 = pd.concat([df1, df2], axis = 1)
df3
```

# Organizando DataFrame (Sort)

```python
data = [[1,2,3], [4,5,6],[7,8,9]]
```

```python
data
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
```

```python
df
```

```python
df.sort_index()
df.sort_index(axis = 1)
df
```

```python
df.sort_index(inplace = True)
df
```

```python
df.sort_index(inplace = True, axis = 1)
df
```

```python
df.sort_values(by = 'X', inplace = True)
df
```

```python
df.sort_values(by = '3', axis = 1, inplace = True)
df
```
