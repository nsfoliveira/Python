# Métodos de Interpolação

```python
data = [0.5, None, None, 0.54, 0.52, None, 0.59, 0.6, None, 0.7]
```

```python
s = pd.Series(data)
s
```

```python
s.fillna(0) # ele substitui todos os nulos por 0 ou o valor que informarmos
```

```python
s.fillna(method='ffill') # ele começa de cima para baixo e pega o valor anterior que seja válido
```

```python
s.fillna(method='bfill') # ele faz a mesma coisa que o ffill faz, mas de baixo para cima
```

```python
s.fillna(s.mean()) # ele preenche os nulos com a média dos outros valores
```

```python
s
```

```python
s.fillna(method='ffill', limit = 1) # limitamos para que ele faça a interpolação uma unica vez
```

# Criando faixa de valor

```python
dados = pd.read_csv('aluguel_v5.csv', sep = ';')
```

```python
dados
```

```python
# As faixas serão categorizados pela quantidade de quartos

# 1 a 2
# 3 a 4
# 5 e 6
# 7 ou mais

classes = [0, 2, 4, 6, 100] # informo o valor minimo e os valores máximos de cada classe
```

```python
quartos = pd.cut(dados.Quartos, classes)
quartos
```

```python
# colocando em uma tabela de frequenta (observação o que '(' não entra na frequencia ']' entra)
pd.value_counts(quartos)
```

```python
labels = ['De 1 a 2 quartos', 'De 3 a 4 quartos', 'De 5 a 6 quartos', 'Acima de 7 quartos']
```

```python
quartos = pd.cut(dados.Quartos, classes, labels = labels)
pd.value_counts(quartos)
```

```python
quartos = pd.cut(dados.Quartos, classes, labels = labels, include_lowest=True) #inclui os menores valores, nesse caso quartos com valor 0
pd.value_counts(quartos)
```

# Mais sobre gráficos

```python
# usando matplotlib para visualizações gráficas
# config para mostrar as visualizações no jupyter 

# inline mostrara um gráfico estática na tela
%matplotlib inline

import matplotlib.pyplot as plt

# configura o tamanho que queremos que essa figura fique na tela
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('E:/Workspace_Natalia/Python/ALURA_F_Python/F_PythonDS_03/aluguel_v6.csv', sep = ';')
```

```python
dados
```

```python
# vamos colocar 4 gráficos dentro dessa figura

area = plt.figure()
```

```python
g1 = area.add_subplot(2,2,1) # 2 linhas, 2 colunas e esse grafico ficara na posição 1
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
 # gráfico de dispersão
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor vs Área')

# gráfico histograma
g2.hist(dados.Valor)
g2.set_title('Histograma Valor')

# Amostra aleatoria dentro do DF
dados_g3 = dados.Valor.sample(100) # dentro do nosso DF, na variável Valor ele selecionara 100 valores para compor a amostra
dados_g3.index = range(dados_g3.shape[0]) # o indice sera refeito, pois a amostra vem fora de ordem, o que pode prejudicar o gráfico
g3.plot(dados_g3)
g3.set_title('Amostra Valor')

# gráfico de barra
grupo = dados.groupby('Tipo')['Valor'] # vai agrupar pelo tipo da variável valor
x = grupo.mean().index
valores = grupo.mean().values
g4.bar(x, valores)
g4.set_title('Valor Médio por Tipo')
```
